```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_probability():
    # Define the transition matrix P
    P = np.array([
        [0.1640, 0.3840, 0.1160, 0.3360],
        [0.0943, 0.3030, 0.1408, 0.4620],
        [0.3120, 0.1995, 0.3030, 0.1855],
        [0.2340, 0.1590, 0.3960, 0.2110]
    ])
    
    # Compute P^3 (three-step transition matrix)
    P3 = np.linalg.matrix_power(P, 3)
    
    # Initial state vector (Monday, known state is (s,h))
    pi_0 = np.array([0, 0, 1, 0])
    
    # Compute the probability distribution after 3 steps
    pi_3 = np.dot(pi_0, P3)
    
    # Probability that the boss is happy on Friday
    prob_happy_friday = pi_3[2] + pi_3[3]  # P(s,h) + P(h,h)
    
    print("Probability that the boss is happy on Friday (using P^3):", round(prob_happy_friday, 4))
    
    # Alternative approach using P^4 (Monday to Friday directly)
    P4 = np.array([
        [0.2824, 0.2364, 0.2656, 0.2156],
        [0.2158, 0.3194, 0.1815, 0.2833],
        [0.1921, 0.2022, 0.3194, 0.2863],
        [0.1502, 0.2454, 0.2429, 0.3616]
    ])
    
    # Compute the probability distribution after 4 steps
    pi_4 = np.dot(pi_0, P4)
    
    # Probability that the boss is happy on Friday using P^4
    prob_happy_friday_alt = pi_4[2] + pi_4[3]
    
    print("Probability that the boss is happy on Friday (using P^4):", round(prob_happy_friday_alt, 4))
    
    # Plot results
    labels = ['(s,s)', '(h,s)', '(s,h)', '(h,h)']
    x = np.arange(len(labels))
    width = 0.35  # width of bars
    
    fig, ax = plt.subplots()
    rects1 = ax.bar(x - width/2, pi_3, width, label='P^3 (Three Steps)')
    rects2 = ax.bar(x + width/2, pi_4, width, label='P^4 (Four Steps)')
    
    # Labels and formatting
    ax.set_xlabel('States')
    ax.set_ylabel('Probability')
    ax.set_title('State Probabilities After 3 and 4 Steps')
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.legend()
    
    plt.show()

# Run the function
calculate_probability()
```
