```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define a simple 1-step transition matrix (Weather Example)
P = np.array([
    [0.7, 0.2, 0.1],  # Sunny -> [Sunny, Cloudy, Rainy]
    [0.3, 0.4, 0.3],  # Cloudy -> [Sunny, Cloudy, Rainy]
    [0.2, 0.3, 0.5]   # Rainy -> [Sunny, Cloudy, Rainy]
])

# Compute 2-step and 3-step transition matrices using Chapman-Kolmogorov
P_2 = np.linalg.matrix_power(P, 2)  # P^2 (2-step)
P_3 = np.linalg.matrix_power(P, 3)  # P^3 (3-step)

# Labels for states
states = ["Sunny", "Cloudy", "Rainy"]

# Function to plot heatmaps for visualization
def plot_heatmap(matrix, title):
    plt.figure(figsize=(6, 4))
    sns.heatmap(matrix, annot=True, fmt=".3f", cmap="Blues", xticklabels=states, yticklabels=states)
    plt.title(title)
    plt.xlabel("Next State")
    plt.ylabel("Current State")
    plt.show()

# Plot 1-step, 2-step, and 3-step transition matrices
plot_heatmap(P, "1-Step Transition Matrix (P)")
plot_heatmap(P_2, "2-Step Transition Matrix (P)")
plot_heatmap(P_3, "3-Step Transition Matrix (P)")
```
